# AI-Driven API Integration Framework - Example

This notebook demonstrates how to interact with the API endpoints natively.

```python
import requests

base_url = "http://127.0.0.1:8000"
```

### 1. Authentication
First, obtain an OAuth2 Bearer token.

```python
auth_response = requests.post(f"{base_url}/token", data={"username": "testuser", "password": "testpass"})
token = auth_response.json().get("access_token")
headers = {"Authorization": f"Bearer {token}"}
print("Acquired Token:", token)
```

### 2. Process Data (Safe Payload)
Send normal data which the AI should classify as safe.

```python
safe_payload = {
    "transaction_amount": 150.5,
    "user_transaction_frequency": 5.0,
    "hour_of_day": 14.5
}

res = requests.post(f"{base_url}/api/v1/process-data", json=safe_payload, headers=headers)
print(res.json())
```

### 3. Process Data (Anomalous Payload)
Send suspicious data which the AI should classify as requiring action.

```python
anomalous_payload = {
    "transaction_amount": 850.0,
    "user_transaction_frequency": 45.0,
    "hour_of_day": 2.0
}

res = requests.post(f"{base_url}/api/v1/process-data", json=anomalous_payload, headers=headers)
print(res.json())
```
